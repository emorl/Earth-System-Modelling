```python
import folium
import pandas as pd

# Daten
data = {
    'longitude': [8.125, -3.625, 15.875],
    'latitude': [48.125, 38.625, 63.625]
}
df = pd.DataFrame(data)

# Europakarte
m = folium.Map(location=[54, 10], zoom_start=4)

# Punkte
for _, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        icon=folium.Icon(color='red')
    ).add_to(m)

# Karte anzeigen
m
```

```python
!pip install cartopy
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Deine Koordinaten
longitudes = [8.125, -3.625, 15.875]
latitudes = [48.125, 38.625, 63.625]

# Erstelle Karte mit Mercator-Projektion
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection=ccrs.Mercator())

# Setze Kartenausschnitt auf Europa
ax.set_extent([-15, 30, 35, 70], crs=ccrs.PlateCarree())

# Ländergrenzen und Küstenlinien hinzufügen
ax.add_feature(cfeature.BORDERS, linewidth=1)
ax.add_feature(cfeature.COASTLINE)

# Punkte einzeichnen
ax.scatter(longitudes, latitudes, color='red', s=100, transform=ccrs.PlateCarree(), zorder=5)

# Titel
plt.title("Untersuchungsgebiete in Deutschland, Spanien, Schweden")

# Karte anzeigen
plt.show()
```
